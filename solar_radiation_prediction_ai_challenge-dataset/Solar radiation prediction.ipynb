```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
train_data=pd.read_csv('TRAIN.csv')
test_data=pd.read_csv('TEST.csv')
```

```python
train_data.drop(['idx','UNIXTime','Time'],axis=1,inplace=True)
test_data.drop(['idx','UNIXTime','Time'],axis=1,inplace=True)
```

```python
train_data['Data']=pd.to_datetime(train_data['Data'])
test_data['Data']=pd.to_datetime(test_data['Data'])
```

```python
import datetime as dt
```

```python
train_data['DayOfWeek']=train_data['Data'].dt.day_name()
test_data['DayOfWeek']=test_data['Data'].dt.day_name()
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
train_data['HoursSunWasAvailable']=pd.to_datetime(train_data['TimeSunSet'])-pd.to_datetime(train_data['TimeSunRise'])
test_data['HoursSunWasAvailable']=pd.to_datetime(test_data['TimeSunSet'])-pd.to_datetime(test_data['TimeSunRise'])
```

```python
train_data['HoursSunWasAvailable']=train_data['HoursSunWasAvailable'].apply(lambda x:str(x).split(' ')[2])
test_data['HoursSunWasAvailable']=test_data['HoursSunWasAvailable'].apply(lambda x:str(x).split(' ')[2])
```

```python
train_data['HoursSunWasAvailable']=train_data['HoursSunWasAvailable'].apply(lambda x:x.split(':')[0])
test_data['HoursSunWasAvailable']=test_data['HoursSunWasAvailable'].apply(lambda x:x.split(':')[0])
```

```python
train_data['HoursSunWasAvailable']=train_data['HoursSunWasAvailable'].astype(int)
test_data['HoursSunWasAvailable']=test_data['HoursSunWasAvailable'].astype(int)
```

```python
train_data.drop(['Data','TimeSunSet','TimeSunRise'],axis=1,inplace=True)
test_data.drop(['Data','TimeSunSet','TimeSunRise'],axis=1,inplace=True)
```

```python
dummies1=pd.get_dummies(train_data['DayOfWeek'],drop_first=True)
dummies2=pd.get_dummies(test_data['DayOfWeek'],drop_first=True)
```

```python
train_data=train_data.join(dummies1)
test_data=test_data.join(dummies2)
```

```python
train_data.drop('DayOfWeek',axis=1,inplace=True)
test_data.drop('DayOfWeek',axis=1,inplace=True)
```

```python
X_train=train_data.copy()
```

```python
X_train.head()
```

```python
X_train.drop('Radiation',axis=1,inplace=True)
```

```python
y_train=train_data['Radiation']
```

```python
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor()
reg.fit(X_train,y_train)
```

```python
reg.score(X_train,y_train)
```

```python
y_pred=reg.predict(test_data)
```
